```python
import pandas as pd
import numpy as np
import pickle
import tabulate
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('dataset.csv')

with open('my_list.pkl', 'rb') as f:
    len_list = pickle.load(f)
```

```python
dataset.describe()
```

```python
print("Features for \",\"")
dataset[dataset["punc"] == ","].describe()[:"mean"]
```

```python
print("Features for \".\"")
dataset[dataset["punc"] == "."].describe()[0:]
```

```python
print("Features for \"!\"")
dataset[dataset["punc"] == "!"].describe()[0:]
# print("Features for \"!\"")
# subset = dataset[dataset["punc"] == "!"]
# subset.describe().loc[["mean","max","min"]]
```

```python
print("Features for no punctuation")
dataset[dataset["punc"] == " "].describe()[0:]
```

```python
subset = dataset[dataset["punc"] == "!"]
x1=subset.describe().loc[["mean"]].reset_index(drop="true")
subset = dataset[dataset["punc"] == "."]
x2=subset.describe().loc[["mean"]].reset_index(drop="true")
subset = dataset[dataset["punc"] == ","]
x3=subset.describe().loc[["mean"]].reset_index(drop="true")
subset = dataset[dataset["punc"] == " "]
x4=subset.describe().loc[["mean"]].reset_index(drop="true")
```

```python
result_df = pd.concat([x1, x2, x3, x4], keys=['!', '.', ',', ' '])
df=result_df.drop(columns=["Time Stamps (s)","Values"])
result_df = df.T

result_df.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Mean Values for Each Punctuation Mark')
plt.xlabel('Columns')
plt.ylabel('Mean Value')
plt.legend(title='Punctuation', bbox_to_anchor=(1.05, 1), loc='upper left', labels=['!', '.', ',', ' '])
plt.grid()
plt.show()
```
