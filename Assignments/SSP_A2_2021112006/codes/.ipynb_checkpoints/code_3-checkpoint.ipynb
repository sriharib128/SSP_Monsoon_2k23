```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_file = "../wavs/q3_audio.wav"
y, sr = librosa.load(audio_file)
print(f"sampling rate = {sr} and no of sample = {len(y)}")
print(f"duration = {len(y)/sr}")
print(f"frames in 20ms = {20*sr/1000}")
```

```python
import numpy as np

def short_fft(y, sr, window_dur, window_shape, shift_dur):
    i = 0
    short_ffts = []

    # Calculate frame length, shift length, and FFT length
    frame_length = int(np.round((sr * window_dur) / 1000))
    shift_length = int(np.round((sr * shift_dur) / 1000))
    fft_length = frame_length  # Set FFT length equal to frame length
#     print(frame_length)

    while i < len(y):
        frame = y[i:i + frame_length]

        # Apply the Hann window by default when window_shape is 'hann'
        if window_shape == 'hann':
            window = np.hanning(frame_length)
        elif window_shape == 'rect':
            window = np.ones(frame_length)
        
        if(len(frame)<frame_length):
            frame = np.concatenate((frame,np.zeros(frame_length-len(frame))),axis=0)
        
        frame = frame * window  # Apply the window function
        fft = np.abs(np.fft.fft(frame))
        short_ffts.append(fft[:(frame_length // 2)+1])

        i = i + shift_length

    short_ffts = np.array(short_ffts).T
    short_ffts = short_ffts / np.max(short_ffts)  # Normalize the FFT

    return short_ffts

window_dur = 20
window_shape = "hann"
shift_dur = 10
stft_inbuilt = short_fft(y,sr,window_dur,window_shape,shift_dur)

# magnitude to dB scale
db_spectrogram = 20 * np.log10((stft_inbuilt+0.0000001)/np.max(stft_inbuilt))

# Create a time vector for the x-axis
num_frames = db_spectrogram.shape[1]
time_vector = np.arange(0, num_frames) * (shift_dur / 1000.0)

# Plot the spectrogram
plt.figure()
plt.imshow(db_spectrogram, cmap='viridis', origin='lower', aspect='auto', extent=[0, time_vector[-1], 0, sr / 2])
plt.colorbar(format="%+2.0f dB")
plt.title(f"Spectrogram - Window Shape: {window_shape}")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.show()
```
