```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import scipy
import aubio
from scipy.signal import find_peaks
```

```python
def num_frames(y, sr, window_dur,shift_dur):
    i = 0
    frames = []
    t =np.array(range(len(y)))/sr
    # print(len(y),len(t))
    frame_length = int(np.round((sr * window_dur) / 1000))
    shift_length = int(np.round((sr * shift_dur) / 1000))
    times =[]
    while i < len(y):
        frame = y[i:i + frame_length]
        times.append((t[i:i+frame_length]))
        i = i + shift_length
        frames.append(frame)
    return frames,times
```

```python
audio_file = "../wavs/q3_audio.wav"
y, sr = librosa.load(audio_file)
frames,times = num_frames(y,sr,(512/sr)*1000,(256/sr)*1000)
print(len(frames),len(frames[1]))
```

```python
t =np.array(range(len(y)))/sr
fig,ax = plt.subplots(figsize = (25,4))
ax.plot(t,y);
i_even = 0
i_odd = 0
for i in range(len(frames)):
    if( i & 1):
        if(i_even):
            temp = "solid"
        else:
            temp = "dotted"
        ax.hlines(y=[-0.4],xmin = times[i][0],xmax = times[i][-1],linestyles=temp)
        ax.text(x=(times[i][0] + times[i][-1]) / 2, y=-0.4+0.05, s=str(i), ha='center', va='center')
        i_even = not (i_even)
    else:
        if(i_odd):
            temp = "solid"
        else:
            temp = "dotted"
        i_odd = not (i_odd)
        ax.hlines(y=[0.5],xmin = times[i][0],xmax = times[i][-1],linestyles=temp)
        ax.text(x=(times[i][0] + times[i][-1]) / 2, y=0.5-0.05, s=str(i), ha='center', va='center')
ax.text(x = -0.03,y = -0.4+0.15,s=f"frame number",ha = 'center',va ='center');
ax.text(x = -0.03,y = 0.4-0.05,s=f"frame number",ha = 'center',va ='center');
ax.set_xlabel("time (s) -->")
ax.set_ylabel("amplitude -->")
ax.set_title("Splitting into frames");
        
```

```python
t =np.array(range(len(y)))/sr
fig,ax = plt.subplots(figsize = (25,4))
voiced_Frame = y[4000:4512]
temp = voiced_Frame
plt.plot(range(len(temp)),temp);
```

```python
hamming_window = np.hamming(512)
frame   = voiced_Frame*hamming_window
temp = frame
plt.plot(range(len(temp)),temp);
plt.xlabel("n")
plt.ylabel("amplitude")
plt.title("Voiced Frame with Hamming Window");
```

### 3.a Auto-correlation of the frame

```python
autocorr = np.correlate(frame,frame,mode = "full")[len(frame)*2//2:]
plt.plot(autocorr);
plt.title("Autocorrelation of the Voiced Frame")
plt.xlabel("n")
plt.ylabel("magnitude");
peaks,_ = find_peaks(autocorr)
plt.plot(peaks, autocorr[peaks],"ro");
n_delay = (peaks[(np.argmax(autocorr[peaks]))])
print(n_delay)
pitch_period = n_delay*1000/sr
pitch_freq = 1000/pitch_period
print(f"pitch period = {pitch_period:0.2f} ms")
print(f"pitch frequency = {pitch_freq:0.2f} Hz")
```

### 3.b Magnitude spectrum of the frame

```python
frame_fft = np.fft.fft(frame)
frame_fft_shift = abs(np.fft.fftshift(frame_fft))
x_freq = np.arange(-sr/2,sr/2, sr/len(frame))
plt.plot(x_freq,frame_fft_shift);
plt.title("Magnitude Spectrum of the Voiced Frame")
plt.xlabel("Hz")
plt.ylabel("magnitude");

peaks,_ = find_peaks(frame_fft_shift)

plt.plot(x_freq[peaks], frame_fft_shift[peaks],"ro");
pitch_freq = (x_freq[peaks[(np.argmax(frame_fft_shift[peaks]))]])
pitch_period=1000/pitch_freq
print(f"pitch period = {pitch_period:0.2f} ms")
print(f"pitch frequency = {pitch_freq:0.2f} Hz")
```

### 3.c LP spectrum of the frame

```python
order = 10
lp_coefficients = librosa.lpc(frame, order=order)
lp_spectrum = np.abs(np.fft.fft(lp_coefficients, n=len(frame)))#[len(frame)//2:]
x_freq = np.arange(0,sr/2, sr/len(frame))

lp_fft_shift = np.fft.fftshift(lp_spectrum)[len(frame)//2:]
plt.plot(x_freq,lp_fft_shift);
plt.title("Magnitude of LP Spectrum of the Voiced Frame")
plt.xlabel("Hz")
plt.ylabel("magnitude");

peaks,_ = find_peaks(lp_fft_shift)
plt.plot(x_freq[peaks], lp_fft_shift[peaks],"ro");
pitch_freq = (x_freq[peaks[0]])
pitch_period=1000/pitch_freq
print(f"pitch period = {pitch_period:0.2f} ms")
print(f"pitch frequency = {pitch_freq:0.2f} Hz")
```

##### Which method provides best estimate
 - In general, the auto-correlation method is the most robust pitch estimation method, but it can be less accurate for unvoiced speech.
 - The magnitude spectrum method is less robust to noise, but it can be more accurate for voiced speech.
 - The LP spectrum method is a compromise between the two, and it is often used in speech recognition applications.
