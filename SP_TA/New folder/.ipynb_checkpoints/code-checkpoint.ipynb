```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
import soundfile as sf
```

```python
def butter_lowpass_filter(data, cutoff, fs, order=12):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = lfilter(b, a, data)
    return y
```

```python
def butter_highpass_filter(data, cutoff, fs, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    y = lfilter(b, a, data)
    return y
```

```python
def match_length(x1, target_length):
    current_length = len(x1)
    
    # Calculate the number of repetitions needed to match the target length
    repetitions = -(-target_length // current_length)  # Equivalent to ceil(target_length / current_length)
    
    # Repeat x1 to match the target length
    matched_x1 = np.tile(x1, repetitions)[:target_length]
    
    return matched_x1
```

```python
def plot_stft_with_lowpass(audio_file, title, ax, cutoff_frequency=None):
    # Load the audio file with a specific sampling rate
    y, sr = librosa.load(audio_file, sr=44100)
    y=match_length(y,1976621)
    
    if cutoff_frequency != None:
        # for music.wav
        y_filt = butter_lowpass_filter(y, cutoff_frequency, sr)
        sf.write(f"{title}_lp.wav", y_filt,sr) # audiowrite
    else:
        # for noise files
        y_highpass = butter_highpass_filter(y,8000, sr)
        y_filt = y + 5*y_highpass
        sf.write(f"{title}_hp.wav", y_filt,sr) # audio write
        
    # Calculate the STFT for the filtered signal
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y_filt)), ref=np.max)

    # Display the STFT
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log', ax=ax)
    ax.set_title(f"{title}=>{len(y)}")
    return y_filt
```

```python
# List of audio files
audio_files = [
    "./ceiling-fan.mp3",
    "./city-traffic.mp3",
    "./pressure-cooker.mp3",
    "./water-pump.wav",
    "./Recording.wav"
]

# Create a 2x2 subplot for the first five audio files
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("STFT Subplots with Lowpass Filtering")

filtered=[]
for i in range(4):
    title = audio_files[i].split("/")[-1].split(".")[0]  # Extract the title from the file path
    temp = plot_stft_with_lowpass(audio_files[i], title, axs[i // 2, i % 2])
    filtered.append(temp)
fig.tight_layout()
plt.show()
```

```python
# Plot "Recording.wav" separately
plt.figure(figsize=(8, 4))
title = audio_files[-1].split("/")[-1].split(".")[0]
music = plot_stft_with_lowpass(audio_files[-1], title, plt.gca(),8000)
plt.show()
```

```python
def addition(t):
    plt.figure(figsize=(12,4))
    
    # Adding singal
    x = music + (filtered[t]) * max(music) / max(filtered[t])
    
    D = librosa.amplitude_to_db(np.abs(librosa.stft(x)), ref=np.max)
    plt.subplot(1, 3, 1)
    librosa.display.specshow(D, sr=44100, x_axis='time', y_axis='log')
    plt.title("Original")

    # Moving Average
    mv_x = np.zeros(len(x))
    mv_x[0] = x[0] / 2
    for i in range(1, len(x)):
        mv_x[i] = ((x[i] + x[i - 1]) / 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(mv_x)), ref=np.max)
    plt.subplot(1, 3, 2)
    librosa.display.specshow(D, sr=44100, x_axis='time', y_axis='log')
    plt.title("Moving Average")

    # Differentiated Signal
    diff_x = np.zeros(len(x))
    diff_x[0] = x[0] / 2
    for i in range(1, len(x)):
        diff_x[i] = ((x[i] - x[i - 1]) * 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(diff_x)), ref=np.max)
    plt.subplot(1, 3, 3)
    librosa.display.specshow(D, sr=44100, x_axis='time', y_axis='log')
    plt.title("Differentiated")
    
    plt.suptitle(f"music + {audio_files[t]}")
    plt.tight_layout()
    plt.show()
```
