```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
import soundfile as sf
```

```python
def butter_lowpass_filter(data, cutoff, fs, order=12):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = lfilter(b, a, data)
    return y
```

```python
def plot_stft_with_lowpass(audio_file, title, ax, cutoff_frequency=None):
    # Load the audio file with a specific sampling rate
    y, sr = librosa.load(audio_file, sr=44100)

    if cutoff_frequency != None:
        y_lowpass = butter_lowpass_filter(y, cutoff_frequency, sr)
        sf.write
        sf.write(f"{title}_lp.wav", y_lowpass,sr)
    else:
        y_lowpass = y
        
    # Calculate the STFT for the filtered signal
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y_lowpass)), ref=np.max)

    # Display the STFT
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log', ax=ax)
    ax.set_title(f"{title}=>{len(y)}")
```

```python
# List of audio files
audio_files = [
    "./ceiling-fan.mp3",
    "./city-traffic.mp3",
    "./pressure-cooker.mp3",
    "./water-pump.wav",
    "./Recording.wav"
]

# Create a 2x2 subplot for the first five audio files
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("STFT Subplots with Lowpass Filtering")

for i in range(4):
    title = audio_files[i].split("/")[-1].split(".")[0]  # Extract the title from the file path
    plot_stft_with_lowpass(audio_files[i], title, axs[i // 2, i % 2])

fig.tight_layout()
plt.show()
```

```python
# Plot "Recording.wav" separately
plt.figure(figsize=(8, 4))
title = audio_files[-1].split("/")[-1].split(".")[0]
plot_stft_with_lowpass(audio_files[-1], title, plt.gca(),8000)
plt.show()
```
